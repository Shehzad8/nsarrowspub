```python
read = pd.read_excel("/gdrive/My Drive/GCPL_Catalog_Hygiene_AmazonFBA_DigitalPanda.xlsx", sheet_name = "December2021_1",header=1)
```

```python
labels = read.iloc[:, 2].tolist()
```

```python
type(labels)
```

```python
new_url = []
labels.pop(0)
for w in labels:
  url = "https://www.amazon.in/dp/"+w
  new_url.append(url)
```

```python
len(new_url)
new_url
```

```python
def product(labels):
  count = 0
  final_list = []
  for i in new_url:
    products_dict = dict()
    s = HTMLSession()
    g = s.get(i)
    
    try:
      ratings =  g.html.xpath('//*[@id="acrPopover"]/span[1]/a/i[1]', first = True).text,
    except:
      ratings = "invalid url for ratings"
    try:
      Product_availability =  g.html.xpath('//*[@id="availability"]/span', first = True).text
    except:
      Product_availability =  "invalid url for availability"
    products_dict['ASIN'] = labels[count]
    products_dict['new_url'] = i
    products_dict['ratings'] = ratings
    products_dict['product_availability'] = Product_availability
    
    count=count+1
    #print(str(count)+ " " ,products_dict)
    final_list.append(products_dict)
  return final_list
```

```python
df1 = product(labels)
```

```python
print(df1)
```

```python
fd = pd.DataFrame(df1)
fd
#fd.to_csv('df1.csv',index = False)
```
